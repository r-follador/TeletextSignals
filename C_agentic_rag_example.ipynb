# Agentic RAG
Slower but more flexibel. Much more dependent on the quality of the LLM.

- LLM Agent queries the documents (i.e. rewrites queries as it seems fit)
- Can query multiple times if necessary
- Needs more LLM calls

*Drawbacks*:
- If the question is too broad (i.e. too many documents are returned), the context gets overwhelmed. The **top_k** and **max_content_chars** params need to be tweaked. See, for example, the Venezuela query, that misses recent key events.

```python
import warnings
from tqdm.std import TqdmWarning

# Silence upstream warnings that are noisy in this notebook environment.
warnings.filterwarnings(
    "ignore",
    message="Core Pydantic V1 functionality isn't compatible with Python 3.14 or greater.",
    category=UserWarning,
    module="langchain_core._api.deprecation",
)

from modules.langchain_agentic_rag import make_teletext_agentic_rag
from modules.notebook_helpers import display_cited_sources, display_answer

rag = make_teletext_agentic_rag(
    model="qwen3.5:4b",
    base_url="http://localhost:11434",
    k=40,
    top_k=15,
    debug=False,
    max_context_chars=10000,
)
def query(question):
    answer, sources = rag.ask_with_sources(question)
    display_answer(answer)
    display_cited_sources(answer, sources)
```

```python
query("What's happening between Venezuela and the US?")
```

```python
query("Wie hoch ist der Umsatz der Migros im Laufe der Jahre?")
```
